```python
import glog
from qiskit import QuantumCircuit, Aer, execute
```

```python
# search for 01

NUM_QUBITS = 2
```

```python
def oracle():
  qc = QuantumCircuit(NUM_QUBITS)
  qc.z(1)
  qc.cz(0, 1)
  gate = qc.to_gate()
  gate.name = "oracle"
  return gate
```

```python
def h_all(qc):
  for i in range(qc.num_qubits):
    qc.h(i)
```

```python
def reflection():
  qc = QuantumCircuit(NUM_QUBITS)
  for i in range(NUM_QUBITS):
    qc.h(i)
    qc.z(i)
  qc.cz(0, 1)
  h_all(qc)
  gate = qc.to_gate()
  gate.name = "reflection"
  return gate
```

```python
grover_circ = QuantumCircuit(NUM_QUBITS, NUM_QUBITS)
h_all(grover_circ)
grover_circ.append(oracle(), [0, 1])
grover_circ.append(reflection(), [0, 1])
grover_circ.measure([0, 1], [1, 0]) # qiskit has bits the opposite way
grover_circ.draw()
```

```python
backend = Aer.get_backend("qasm_simulator")
sim = execute(grover_circ, backend, shots = 1024)
glog.info(sim.result().get_counts())
```
