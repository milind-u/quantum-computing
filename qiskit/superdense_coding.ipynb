```python
from qiskit import QuantumCircuit, Aer, execute
```

```python
def gen_bell_pair(qc):
  qc.h(0)
  qc.cx(0, 1)
```

```python
def encode(qc, msg):
  if msg[1]:
    qc.z(1)
  if msg[0]:
    qc.x(1)
```

```python
def decode(qc):
    qc.cx(1, 0)
    qc.h(1)
```

```python
for i in range(4):
  msg = [i >= 2, i % 2]
  print("msg: %u%u" % (msg[0], msg[1]))

  qc = QuantumCircuit(2, 2)
  gen_bell_pair(qc)
  encode(qc, msg)
  decode(qc)
  qc.measure([0, 1], [1, 0])

  backend = Aer.get_backend("qasm_simulator")
  counts = execute(qc, backend, shots = 1024).result().get_counts()

  print("counts: %s\n" % counts)
```
